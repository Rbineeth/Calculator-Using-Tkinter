```python
from tkinter import *
import math

window = Tk()
window.title("Calculator by R Bineeth")
window.geometry("400x600")

def on_button_click(event):
    text = event.widget.cget("text")
    current_text = entry1.get()
    
    if text == '=':
        try:
            result = str(eval(current_text))
            entry1.delete(0, END)
            entry1.insert(END, result)
        except Exception as e:
            entry1.delete(0, END)
            entry1.insert(END, "Error")
    
    elif text == 'C':
        entry1.delete(0, END)

    elif text == 'CE':
        entry1.delete(0, END)
    
    elif text == '⌫':  
        current_text = entry1.get()
        entry1.delete(0, END)
        entry1.insert(END, current_text[:-1])

    elif text == '1/x':
        try:
            result = str(1 / float(current_text))
            entry1.delete(0, END)
            entry1.insert(END, result)
        except:
            entry1.delete(0, END)
            entry1.insert(END, "Error")

    elif text == 'x²':
        try:
            result = str(float(current_text) ** 2)
            entry1.delete(0, END)
            entry1.insert(END, result)
        except:
            entry1.delete(0, END)
            entry1.insert(END, "Error")

    elif text == '√x':
        try:
            result = str(math.sqrt(float(current_text)))
            entry1.delete(0, END)
            entry1.insert(END, result)
        except:
            entry1.delete(0, END)
            entry1.insert(END, "Error")

    elif text == '%':
        try:
            result = str(float(current_text) / 100)
            entry1.delete(0, END)
            entry1.insert(END, result)
        except:
            entry1.delete(0, END)
            entry1.insert(END, "Error")

    elif text == '+/-':
        try:
            result = str(-float(current_text))
            entry1.delete(0, END)
            entry1.insert(END, result)
        except:
            entry1.delete(0, END)
            entry1.insert(END, "Error")

    else:
        entry1.insert(END, text)

entry1 = Entry(window, font=("Arial", 30), justify="right", borderwidth=10)
entry1.pack()

button_labels = [
    '%', 'CE', 'C', '⌫',
    '1/x', 'x²', '√x', '/',
    '7', '8', '9', '*',
    '4', '5', '6', '-',
    '1', '2', '3', '+',
    '+/-', '0', '.', '='
]

frame = Frame(window, bg='white')
frame.pack()

i = 0
for label in button_labels:
    button = Button(frame, text=label, font=("Arial", 20), padx=15, pady=15, bg='gray', fg='black')
    button.grid(row=i//4, column=i%4, padx=10, pady=10, sticky="nsew")
    button.bind("<Button-1>", on_button_click)
    i += 1


for row in range(6):
    frame.grid_rowconfigure(row, weight=1)
for col in range(4):
    frame.grid_columnconfigure(col, weight=1)

window.mainloop()


    


```
